```python
import matplotlib.pyplot as plt
```

```python


data = {
    "Time (msec)": [(1, 80.000000, 0.019073), (21, 40.000000, 0.038147), (41, 30.000000, 0.050863), (61, 40.000000, 0.038147),
                   (81, 40.000000, 0.038147), (101, 30.000000, 0.050863), (121, 40.000000, 0.038147), (141, 50.000000, 0.030518),
                   (161, 40.000000, 0.038147), (181, 40.000000, 0.038147)],
    "Rate (MB/s)": [(1, 0.019073), (21, 0.038147), (41, 0.050863), (61, 0.038147), (81, 0.038147), (101, 0.050863), (121, 0.038147),
                    (141, 0.030518), (161, 0.038147), (181, 0.038147)]
}

# Extracting data for plotting
x = [item[0] for item in data['Time (msec)']]
y1 = [item[1] for item in data['Time (msec)']]
y2 = [item[1] for item in data['Rate (MB/s)']]

# Plotting Time (msec)
plt.figure(figsize=(10, 6))
plt.plot(x, y1, marker='o', linestyle='-', color='b', label='Time (msec)')

# Adding labels and title for Time (msec)
plt.xlabel('Number of Blocks')
plt.ylabel('Time (msec)')
plt.title('Time (msec) vs Number of Blocks')
plt.legend()
plt.show()

# Plotting Rate (MB/s)
plt.figure(figsize=(10, 6))
plt.plot(x, y2, marker='o', linestyle='-', color='r', label='Rate (MB/s)')

# Adding labels and title for Rate (MB/s)
plt.xlabel('Number of Blocks')
plt.ylabel('Rate (MB/s)')
plt.title('Rate (MB/s) vs Number of Blocks')
plt.legend()
plt.show()
```
